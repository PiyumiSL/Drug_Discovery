```python
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/compound list.xlsx')
df.head()
```

```python
df = df.rename(columns={'YOUR_COLUMN_NAME': 'Molecule ChEMBL ID'})
```

```python
df = df.rename(columns={'YOUR_COLUMN_NAME': 'CHEMBL_ID'})
```

```python
from chembl_webresource_client.new_client import new_client

molecule = new_client.molecule

def fetch_smiles(chembl_id):
    try:
        res = molecule.get(chembl_id)
        return res['molecule_structures']['canonical_smiles']
    except:
        return None

df['SMILES'] = df['Molecule ChEMBL ID'].apply(fetch_smiles)
df.head()
```

```python
from rdkit import Chem
df['mol'] = df['SMILES'].apply(lambda x: Chem.MolFromSmiles(x) if x is not None else None)
```

```python
df.dropna(subset=['SMILES', 'mol'], inplace=True)
print("Rows with missing SMILES or mol objects have been dropped.")
print("New DataFrame shape:", df.shape)
display(df.head())
```

```python
missing_smiles = df['SMILES'].isnull().sum()
missing_mol_objects = df['mol'].isnull().sum()

print(f"Number of missing SMILES strings: {missing_smiles}")
print(f"Number of missing RDKit molecule objects: {missing_mol_objects}")

if missing_smiles > 0 or missing_mol_objects > 0:
    print("\nRows with missing SMILES or RDKit molecule objects:")
    display(df[df['SMILES'].isnull() | df['mol'].isnull()])
```

```python
from rdkit import Chem
df['mol'] = df['SMILES'].apply(lambda x: Chem.MolFromSmiles(x) if x is not None else None)

print("SMILES strings have been converted to RDKit Molecule objects.")
display(df.head())
```

```python
from rdkit.Chem import Descriptors, rdMolDescriptors

df['MolWt'] = df['mol'].apply(lambda x: Descriptors.MolWt(x))
df['MolLogP'] = df['mol'].apply(lambda x: Descriptors.MolLogP(x))
df['NumHAcceptors'] = df['mol'].apply(lambda x: Descriptors.NumHAcceptors(x))
df['NumHDonors'] = df['mol'].apply(lambda x: Descriptors.NumHDonors(x))
df['NumRotatableBonds'] = df['mol'].apply(lambda x: Descriptors.NumRotatableBonds(x))
df['FractionCSP3'] = df['mol'].apply(lambda x: rdMolDescriptors.CalcFractionCSP3(x))
df['TPSA'] = df['mol'].apply(lambda x: Descriptors.TPSA(x))

print("RDKit descriptors have been calculated and added to the DataFrame.")
display(df.head())
```

```python
output_path = '/content/drive/MyDrive/Colab Notebooks/chembl_descriptors.csv'
df.to_csv(output_path, index=False)
print(f"Final descriptor table saved to: {output_path}")
```
